```python
import pandas as pd

dfs = []

for year in range(2016, 2023):
    for quarter in range(1, 5):
        try:
            df = pd.read_csv(f'diaries/diary{str(year)[2:]}/fmld{year % 100}{quarter}.csv')
            
            df['quarter'] = quarter
            df['year'] = year
            
            dfs.append(df)
        except Exception as e:
            print(f"An error occurred while reading diary{str(year)[2:]}/fmld{year % 100}{quarter}.csv: {e}")

combined_df = pd.concat(dfs, ignore_index=True)

combined_df.columns = combined_df.columns.str.lower()
df = combined_df
```
