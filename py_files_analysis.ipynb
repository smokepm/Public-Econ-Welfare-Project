```python
df['during_post_covid_dummy'] = (((df['year'].isin([2020]) & df['quarter'].between(2, 4)) | 
    (df['year'].isin([2021, 2022]))).astype(int))

df[df['during_post_covid_dummy'] == 1]
```

```python
reg = smf.ols(formula = "expen ~ age_ref + male + white + black + american_indian + asian + other + multi_race + during_post_covid_dummy + welf_reic + fwagex", data=df)
results = reg.fit()
print(results.summary())
```
