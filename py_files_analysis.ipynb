```python
df['during_post_covid_dummy'] = (((df['year'].isin([2020]) & df['quarter'].between(2, 4)) | 
    (df['year'].isin([2021, 2022]))).astype(int))

print(df[df['during_post_covid_dummy'] == 1])
```

```python
df.loc[df['sex_ref']==1, 'male']=1
df.loc[df['sex_ref']==2, 'male']=0

df.head(5)
```

```python
reg = smf.ols(formula = "expen ~ age_ref + male + white + black + american_indian + asian + other + multi_race + during_post_covid_dummy + welf_reic + fwagex", data=df)
results = reg.fit()
print(results.summary())
```

```python
X = df[['age_ref', 'sex_ref', 'white', 'black', 'american_indian',
       'asian', 'other', 'multi_race', 'during_post_covid_dummy', 'welf_reic', 'fwagex']]
y = df['expen']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_, X.columns, columns=['Coef'])
coeff_df
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test, predictions, color='grey')
```

```python
sns.histplot((y_test-predictions), bins=50, color='grey')
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
